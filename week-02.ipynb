Mari kita bahas pertanyaan-pertanyaan Anda satu per satu:

1. **Perbedaan Gambar Tanpa dan Dengan Matplotlib:**
   - Matplotlib adalah library Python yang  digunakan untuk visualisasi data dalam bentuk grafik dan plot, termasuk plot gambar (images).
   - Jika menampilkan image tanpa Matplotlib, biasanya  hanya akan melihat file gambar tersebut dalam tampilan standar, seperti yang diberikan oleh sistem kita nanti.
   - Namun kalau kita menggunakan Matplotlib untuk show image, nanti kita dapat manipulasi berbagai aspek tampilan gambar, misalnya seperti skala, ukuran, warna latar belakang, dan lainnya.

2. **Perbedaan dan Pengaruh Pembuatan Black Image antara tipe data int16 dan int32:** 
   - `int16` dan `int32` adalah jenis tipe data integer dengan berbeda ukuran. `int16` memiliki 16 bit (2 byte) sementara `int32` memiliki 32 bit (4 byte).
   - Pengaruh utama adalah masalah penyimpnan. Jadi, untuk Gambar dengan tipe data `int16` akan membutuhkan setengah size file (dibandingkan dengan `int32`) untuk menyimpan setiap px-nya. Namun, hal ini juga memperlihatkan bahwa `int16` memiliki rentang nilai yang lebih terbatas dibandingkan dengan `int32`. 

3. **Kegunaan "google.colab.patches import cv2_imshow":**
   - `google.colab.patches` adalah bagian dari library `google.colab`, sebuah environment dari google colab cloud yang digunakan untuk pengembangan dan eksperimen Python.
   - `cv2_imshow` adalah method dari `google.colab.patches` yang digunakan untuk menampilkan gambar dalam notebook Google Colab.
   - Kegunaannya adalah untuk menampilkan gambar dalam notebook Colab dengan antarmuka yang mirip dengan tampilan gambar dalam aplikasi lokal. Kita butuh hal tersebut karena google colab mem-block penggunakan imshow dari bawaan opencv

4. **Kegunaan "skimage import io":**
   - `skimage` adalah modul dalam library Scikit-image, yang digunakan untuk _image processing_ dalam Python.
   - `io` adalah submodul di Scikit-image yang digunakan untuk operasi input-output gambar, seperti membaca atau menulis gambar dari atau ke berbagai format file gambar.

## 5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

```python
import random

width, height = 1080, 1080
blank_image = np.zeros((height, width, 3), np.uint8) * 255
```

```python
for i in range(5):
    x1, y1 = random.randint(0, width-50), random.randint(0, height-50)
    x2, y2 = x1 + random.randint(20, 50), y1 + random.randint(20, 50)
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    thickness = random.randint(1, 5)
    cv.rectangle(blank_image, (x1, y1), (x2, y2), color, thickness)
```

```python
plt.title('Random image')
plt.imshow(blank_image)
```
